## Redes Neurais

### Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

%matplotlib inline
plt.rcParams['figure.dpi'] = 200
plt.rcParams['figure.figsize'] = [15, 15]

#define função para criar plots com fronteiras de decisão
def plot_decision_boundary(clf, X, y):
    x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    h = .02
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    
    Z = Z.reshape(xx.shape)
    plt.figure(1, figsize=(4, 3))
    plt.pcolormesh(xx, yy, Z, cmap=plt.cm.tab20)

    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.tab20b, s=5)
    plt.axis('off')
```

```python
X, y = datasets.make_moons(n_samples=500, random_state=420)
X = X + np.random.normal(0.0, 0.1, X.shape)
plt.scatter(X[:,0], X[:,1], c=y)
```

### Redes Neurais Multicamadas com scikit-learn

http://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html#sklearn.neural_network.MLPClassifier


```python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(solver='sgd',
                    learning_rate_init=0.01,
                    max_iter=600,
                    hidden_layer_sizes=(200,200),
                    random_state=420)

clf.fit(X, y)
z = clf.predict(X)
```

```python
plot_decision_boundary(clf, X, y)
```

## Clustering

https://scikit-learn.org/stable/modules/clustering.html

```python
from sklearn.cluster import KMeans

X, y = datasets.make_blobs(n_samples=1500, centers=3, random_state=420)

y_pred = KMeans(n_clusters=5).fit_predict(X)

#y_pred1 = KMeans(n_clusters=8).fit_predict(X)

#plt.figure()
#plt.scatter(X[:, 0], X[:, 1], c=y)

plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
#plt.figure()
#plt.scatter(X[:, 0], X[:, 1], c=y_pred1)
```

```python
X, y = datasets.make_moons(n_samples=1500, random_state=420)
X = X + np.random.normal(0.0, 0.05, X.shape)

y_pred = KMeans(n_clusters=2, random_state=420).fit_predict(X)

from sklearn.cluster import DBSCAN
y_pred1 = DBSCAN().fit_predict(X)

plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=y_pred)

plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=y_pred1)
```
